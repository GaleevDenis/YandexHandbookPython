```python
# B Решена

class Point:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def move(self, move_x, move_y):
        self.x = self.x + move_x
        self.y = self.y + move_y
        return self.x, self.y
    
    def length(self, tochka):
        if self.x >= 0 and tochka.x >= 0:
            x_min = min(self.x, tochka.x)
            x_max = max(self.x, tochka.x)
            x = x_max - x_min
        elif self.x <= 0 and tochka.x <= 0:
            x_min = min(abs(self.x), abs(tochka.x))
            x_max = max(abs(self.x), abs(tochka.x))
            x = x_max - x_min
        else:
            x_min = min(self.x, tochka.x)
            x_max = max(self.x, tochka.x)
            x = abs(x_min) + x_max
        if self.y >= 0 and tochka.y >= 0:
            y_min = min(self.y, tochka.y)
            y_max = max(self.y, tochka.y)
            y = y_max - y_min
        elif self.y <= 0 and tochka.y <= 0:
            y_min = min(abs(self.y), abs(tochka.y))
            y_max = max(abs(self.y), abs(tochka.y))
            y = y_max - y_min
        else:
            y_min = min(self.y, tochka.y)
            y_max = max(self.y, tochka.y)
            y = abs(y_min) + y_max
        result = round((x ** 2 + y ** 2) ** 0.5, 2)
        return result
    

first_point = Point(2, -7)
second_point = Point(7, 9)
print(first_point.length(second_point))
print(second_point.length(first_point))
```

```python
# I Решена

class Queue:
    
    def __init__(self):
        self.stuck = []
    
    def push(self, item):
        self.stuck.append(item)
        
    def pop(self):
        return self.stuck.pop(0)
    
    def is_empty(self):
        if self.stuck != []:
            return False
        else:
            return True
    
queue = Queue()
for item in ("Hello,", "world!"):
    queue.push(item)
while not queue.is_empty():
    print(queue.pop())
```

```python
# J Решена

class Stack:
    
    def __init__(self):
        self.scteck = []
    
    def push(self, item):
        self.scteck.insert(0, item)
        
    def pop(self):
        return self.scteck.pop(0)
    
    def is_empty(self):
        if self.scteck != []:
            return False
        else:
            return True
        
stack = Stack()
for item in ("Hello,", "world!"):
    stack.push(item)
while not stack.is_empty():
    print(stack.pop())
```

```python
# D решена


class Programmer:

    def __init__(self, name, position, full_time=0, salary=0):
        self.name = name
        self.post = {"Junior": 10, "Middle": 15, "Senior": 20}
        self.position = position
        self.full_time = full_time
        self.salary_per_hour = self.post[self.position]
        self.salary = salary

    def work(self, time):
        self.full_time += time
        self.salary += time * self.salary_per_hour

    def rise(self):
        match self.position:
            case "Junior":
                self.position = "Middle"
            case "Middle":
                self.position = "Senior"
            case "Senior":
                self.post["Senior"] += 1
        self.salary_per_hour = self.post[self.position]

    def info(self):
        return f"{self.name} {self.full_time}ч. {self.salary}тгр."


programmer = Programmer("Васильев Иван", "Junior")
programmer.work(750)
print(programmer.info())
programmer.rise()
programmer.work(500)
print(programmer.info())
programmer.rise()
programmer.work(250)
print(programmer.info())
programmer.rise()
programmer.work(250)
print(programmer.info())
```

```python
from collections.abc import Iterable


class Shape:
    def __init__(self, color, radius):
        self.color = color
        self.radius = 21

    def verni(self, money):
        self.money = "money"

class Circle(Shape):
    def __init__(self, color, radius = 0):
        super().__init__(color, radius)

    def __dir__(self) -> Iterable[str]:
        return super().__dir__()
    
    def verni(self, money):
        super().verni(money=money)
        return self.money

# Создает объект Circle
my_circle = Circle("Красный")
print(my_circle.verni('123'))
print(my_circle.color, my_circle.radius) # Вывод: 5
```
